```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
slices = [50, 150, 300 ]
labels = ['One', 'Two', 'Three']

plt.pie(slices, labels=labels)
plt.title('im pie chart')
```

```python
std_count = [130, 550, 250, 400]
depts = ["DS", "CS", "SE", "IT"]
explode = [0.2, 0, 0, 0]

plt.pie(std_count, labels=depts, explode=explode, shadow=True, autopct='%.1f%%')
```
