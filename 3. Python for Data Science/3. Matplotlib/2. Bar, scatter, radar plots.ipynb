```python
df = pd.read_csv('../datasets/so_survey_subset.csv')
df
```

```python
from collections import Counter
```

```python
df.dropna(inplace=True)
```

```python
langs_df = df['LanguageWorkedWith']
```

```python
langs_counter = Counter(';'.join(langs_df.tolist()).split(';'))
langs_sorted = sorted(langs_counter.items(), key=lambda item: item[1], reverse=True)
```

```python
langs = [l[0] for l in langs_sorted]
papularity = [l[1] for l in langs_sorted]
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.barh(langs, papularity)
```
