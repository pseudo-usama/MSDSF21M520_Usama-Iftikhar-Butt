```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 20)
y = x ** 2
```

```python
plt.plot(x, y)
plt.title('my title')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(range(-2, 3), ['a', 'b', 'c', 'd', 'e'])
plt.yticks(range(0, 4, 2), ['a', 'b'])
plt.suptitle('I\'m figure')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y, label='asdf')
ax2.plot(x, y)

ax1.set_title('ax1')
ax2.set_title('ax2')
ax1.set_xticks(range(-2, 3), ['a', 'b', 'c', 'd', 'e'])

ax1.legend()

ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')

fig.suptitle('figure title')
```

```python
fig = plt.figure()

ax = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

ax.plot(x, y, label='asdf')
ax.legend()

ax.set_title('ax1')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_xticks(range(-2, 3))

fig.suptitle('figure title')
```
